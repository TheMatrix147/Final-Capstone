# Statistical Analysis

```python
# Goal of analysis is to see if a cause-and-effect relationship exists between
# the variables, snoring_rate, respiration_rate, body_temperature, limb_movement,
# blood_oxygen, eye_movement, heart_rate, and the variable, stress_level during sleep.
```

```python
plt.figure(figsize=(16,12))

ax = sns.heatmap(corr.corr(), annot=True, cmap='RdBu', vmin=-1, vmax=1)
```

```python
# Based on the heatmap, we can see that a very strong positive linear relationship exists between stress_level
# and snoring_rate, denoted by the correlation coefficient of 0.98. I will use this linear relationship
# for my linear regression model, with the independent variable being stress_level. 
```

```python
# Null Hypothesis: Stress levels do not affect one's snoring_rate while sleeping.
# Alternate Hypothesis: Stress levels affect one's snoring_rate while sleeping.
```

```python
# set X 

x1 = corr['stress_level']

# set Y 

y = corr['snoring_rate']
```

```python
# Visualize relationship between X and Y
```

```python
plt.figure(figsize=(10, 7))
sns.regplot(x = 'stress_level', y = 'snoring_rate', data = corr, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
```

```python
x = sm.add_constant(x1)
```

```python
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
# R-squared

# The model's r-squared measures the variability of the model's data. Because 
# 0.951 is a higher variability, this indicates that the chosen model is a good fit
# for the data.
```

```python
# Coefficient of Independent Variable

# The model's coefficient of the independent variable (b1) quantified the effect the 
# stress_level variable had on the snoring_rate variable. It looks like the effect
# was quantified by 13.3500.
```

```python
# P-value of T-statistic

# The model's p-value of the t-statistic shows if the input variable is 
# significant or not in relation to the Y variable. The p-value is 0.000, 
# so in this case, we reject the null hypothesis since b1 is significant.
```

```python
# P-value of F-statistic

# The f-statistic tests the overall significance of the model. Thus, if 
# its p-value is less than 0.05, the overall model is significant. The 
# p-value of the model's f-statistic is 0.00, so the overall model is significant.
```
